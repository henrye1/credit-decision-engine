```python

from pydantic import BaseModel, model_serializer, PrivateAttr
from pydantic import SerializerFunctionWrapHandler
from pydantic import SerializationInfo
```

```python
class SerializedConfig(BaseModel):
    key: str
    _loaded_config: dict = PrivateAttr(default_factory=dict)


    @model_serializer(mode="wrap")
    def serialize(self, handler: SerializerFunctionWrapHandler, info: SerializationInfo) -> dict:
        result = handler(self)
        ctx = info.context
        if "decider_versioned_config" in ctx:

            ctx["decider_versioned_config"][self.key] = self._loaded_config
        return result
```

```python
cfg = SerializedConfig(key="qwert",)

cfg._loaded_config = {"test": "test"}
output_cfg = {}
cfg.model_dump(context={"decider_versioned_config": output_cfg})
output_cfg
```
